```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Sales': [10, 25, 40, 30],
    'Expenses': [8, 20, 35, 25]
})

# First graph
plt.scatter(df['Sales'], df['Expenses'], color='silver', label='Expenses')

# Second graph inside same chart
plt.plot(df['Sales'], df['Expenses'], color='silver', marker='o', label='Line')

plt.xlabel("Sales")
plt.ylabel("Expenses")
plt.title("Scatter + Line Graph")
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Sales': [10, 25, 40, 30],
    'Expenses': [8, 20, 35, 25]
})

# Bar Chart
plt.bar(df.index, df['Sales'], color='pink', label='Sales')

# Histogram inside same graph
plt.hist(df['Expenses'], color='blue', alpha=0.5, label='Expenses')

plt.title("Bar + Histogram")
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Sales': [10, 25, 40, 30]
})

# Bar Chart
plt.subplot(1, 2, 1)
plt.bar(df['Category'], df['Sales'], color='grey')
plt.title("Bar Chart")

# Pie Chart
plt.subplot(1, 2, 2)
plt.pie(
    df['Sales'],
    labels=df['Category'],
    colors=['blue', 'green', 'pink', 'purple'],
    autopct='%1.1f%%'
)
plt.title("Pie Chart")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Sales': [10, 20, 30, 40, 50],
    'Profit': [5, 15, 25, 35, 45]
})

# Subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Bar Chart
ax[0].bar(df.index, df['Sales'], color='blue')
ax[0].set_title("Bar Chart")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("Sales")

# Histogram
ax[1].hist(df['Profit'], color='orange', alpha=0.7)
ax[1].set_title("Histogram")
ax[1].set_xlabel("Profit")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Data
df = pd.DataFrame({
    'Sales': [10, 25, 40, 30],
    'Expenses': [8, 20, 35, 25]
})

# Main figure
fig, ax = plt.subplots()

# First graph (line graph)
df.plot(ax=ax)
ax.set_title("Main Graph")

# Second graph (bar graph inside the first)
ax2 = inset_axes(ax, width="40%", height="40%", loc="lower right")

df.plot(kind='bar', ax=ax2)
ax2.set_title("Bar Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
dist_data = np.random.normal(0, 1, 1000) # Histogram data
bar_labels = ['A', 'B', 'C']
bar_values = [15, 30, 45]             # Bar graph data

fig, ax1 = plt.subplots()

# 1. Primary Axis: Histogram
ax1.hist(dist_data, bins=20, color='skyblue', alpha=0.6, label='Frequency')
ax1.set_ylabel('Histogram Count', color='blue')

# 2. Secondary Axis: Bar Graph
ax2 = ax1.twinx()  # Create twin axis sharing the x-axis
# Aligning bars to specific x-coordinates for overlay
ax2.bar([-1, 0, 1], bar_values, width=0.4, color='orange', alpha=0.5, label='Bar Value')
ax2.set_ylabel('Bar Scale', color='red')

plt.title("Histogram and Bar Chart with Different Axes")
plt.show()
```
